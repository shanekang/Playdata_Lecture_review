```
import re
```

```
def cleanText(readData):
  #텍스트에 포함되어 있는 특수 문자 제거
  text = re.sub('[^\w\s]', ' ', readData)
  return text
```

```
cleanText(text)
```

```
import numpy as np
```

```
def preprocess(text):
  text = text.lower()
  text = text.replace('.', ' .')
  text = text.replace('\n', ' ')
  text = cleanText(text)
  words = text.split(' ')

  word_to_id = {}
  id_to_word = {}
  for word in words:
    if word not in word_to_id:
      new_id = len(word_to_id)
      word_to_id[word] = new_id
      id_to_word[new_id] = word

  corpus = np.array([word_to_id[w] for w in words])

  return corpus, word_to_id, id_to_word
```

```
def create_co_matrix(corpus, vocab_size, window_size=1):
  corpus_size = len(corpus)

  # 비어있는 행렬 먼저 선언 
  co_matrix = np.zeros((vocab_size, vocab_size), dtype=np.int32)

  # corpus를 반복문으로 탐색하며 동시발생 행렬 값 추가 
  for idx, word_id in enumerate(corpus):
    for i in range(1, window_size+1):
      left_idx = idx - i
      right_idx = idx + i

      # 각각 왼쪽/오른쪽의 동시발생행렬 계산, 맨 왼쪽, 맨 오른쪽은 제외 
      if left_idx >= 0:
        left_word_id = corpus[left_idx]
        co_matrix[word_id, left_word_id] += 1

      if right_idx < corpus_size:
        right_word_id = corpus[right_idx]
        co_matrix[word_id, right_word_id] += 1

  return co_matrix
```

```
# 분모에 0이 들어가는 것을 방지하기 위해 작은 값 더해줌
def cos_similarity(x, y, eps=1e-8):
  nx = x / (np.sqrt(np.sum(x ** 2)) + eps)
  ny = y / (np.sqrt(np.sum(y ** 2)) + eps)
  return np.dot(nx, ny)
```

```
def most_similar(query, word_to_id, id_to_word, word_matrix, top=5):
  # 검색어를 꺼낸다 (query)
  if query not in word_to_id:
    print('%s is not in words'%query)
    return
  
  print('\n[query] ' + query)
  query_id = word_to_id[query]
  query_vec = word_matrix[query_id]

  # cosine similarity
  vocab_size = len(id_to_word)
  similarity = np.zeros(vocab_size)
  for i in range(vocab_size):
    similarity[i] = cos_similarity(word_matrix[i], query_vec)

  # order by ranking
  count = 0
  for i in (-1 * similarity).argsort():
    if id_to_word[i] ==query:
      continue
    print(' %s: %s' % (id_to_word[i], similarity[i]))

    count += 1
    if count >= top:
      return
```

```
corpus, word_to_id, id_to_word = preprocess(text)
vocab_size = len(word_to_id)
C = create_co_matrix(corpus, vocab_size)

most_similar('손흥민', word_to_id, id_to_word, C, top=5)
```

```
import tensorflow as tf
import numpy as np
```

```
train_set = []
for i in range(1, len(corpus)-1):
  target_id = corpus[i]
  context = [corpus[i-1], corpus[i+1]]
  train_set.append((context, target_id))
```

```
X = np.array([data[0] for data in train_set])
y = np.array([data[1] for data in train_set])
```

```
y = tf.keras.utils.to_categorical(y, num_classes=vocab_size)
```

```
embedding_layer = tf.keras.layers.Embedding(vocab_size, 3, input_shape=(2,))
model = tf.keras.Sequential(
    [
     embedding_layer,
     tf.keras.layers.GlobalAveragePooling1D(),
     tf.keras.layers.Dense(vocab_size, activation='softmax')
    ]
)
```

```
model.compile(optimizer='adam', loss='categorical_crossentropy')
history = model.fit(X, y, batch_size=X.shape[0], epochs=500)
```

```
word_weights = embedding_layer.get_weights()[0]
```

```
# SVD
U, S, V = np.linalg.svd(word_weights)
```
